### Extension substep 3

```python
%cd /content/code

import torch
import torch.nn as nn
from core.models.step_matching import StepMatchingModule

num_visual_steps = 10
num_graph_nodes = 7
visual_dim = 768
text_dim = 256

visual_feats = torch.randn(num_visual_steps, visual_dim, requires_grad=True)
text_feats = torch.randn(num_graph_nodes, text_dim, requires_grad=True)

module = StepMatchingModule(visual_dim=visual_dim, text_dim=text_dim)
```

```python
%cd /content/code

print("Starting substep 3 verification")
print("[TEST 1] Testing Forward Pass and Shapes...")
updated_nodes, row_ind, col_ind = module(visual_feats, text_feats)

# The output nodes must perfectly match the original graph nodes count and dimension
expected_shape = (num_graph_nodes, text_dim)
assert updated_nodes.shape == expected_shape, f"Shape Error! Expected {expected_shape}, got {updated_nodes.shape}"
print("Dimensionality is correct. Output shape is ready for the Graph Neural Network (GNN).")
```

```python
%cd /content/code

print("\n[TEST 2] Testing Hungarian Matching Logic...")
expected_matches = min(num_visual_steps, num_graph_nodes)

# Check if the number of matches is correct
assert len(row_ind) == expected_matches, "Matching count is incorrect."

# Check for strictly 1-to-1 assignments (no duplicate indices allowed)
assert len(set(row_ind)) == len(row_ind) and len(set(col_ind)) == len(col_ind), "Error: Duplicate assignments found! Matching is not 1-to-1."
print(f"Matching logic is solid. Successfully created {len(row_ind)} unique 1-to-1 pairs.")
```

```python
%cd /content/code

print("\n[TEST 3] Testing Gradient Flow (Learnability)...")

# We simulate a target output to compute a dummy loss and trigger backpropagation
dummy_target = torch.randn(num_graph_nodes, text_dim)
criterion = nn.MSELoss()

# Calculate loss and backpropagate
loss = criterion(updated_nodes, dummy_target)
loss.backward()

# Verify that the learnable parameters in the projection heads received gradients
has_gradients = False
for name, param in module.named_parameters():
    if param.grad is not None:
        has_gradients = True
        break

assert has_gradients, "Critical Error: Gradients are NOT flowing. The module cannot learn."
print("Gradients are flowing correctly. The Projection and Fusion heads will update during training.")
print("\n--- All tests passed! Substep 3 is ready for substep 4 integration. ---")
```
